### Load Necessary Libraries

```python
from matplotlib import pyplot as plt #import matplotlib.pyplot as plt    - same instruction
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [1, 2, 3]
y = [2, 4, 6]

#Resize graph
plt.figure(figsize=(5,3), dpi=300) #5x3 inches, dpi - pixels per inch, recommended ratio around 300, in this case 1500x900

#Line 1
#longhand notation:
#plt.plot(x, y, label='2x', color='red', linewidth=2, marker='.', linestyle='--', markersize='10', markeredgecolor='blue')

#Shorthand notation:
#fmt = '[color][marker][line]'
plt.plot(x, y, 'b^--', label='2x')

## Line number 2
x2 = np.arange(0, 4, 0.5) #arrange from 0 to 4 every 0.5
print(x2)

#plot part of the graph as line
plt.plot(x2[:5], x2[:5]**2, 'r', label='X^2') #x2^2 = x2**2, x2[:5] - take the first 5 values

#Plot remainder of graph as dot
plt.plot(x2[4:], x2[4:]**2, 'r--') #from the fifth value onwards, make the line as dash

#Add title with format
plt.title('First Graph', fontdict={'fontname':'Comic Sans MS', 'fontsize':20}) #font list: https://www.youtube.com/redirect?event=video_description&redir_token=QUFFLUhqbEp3eWhxS2RpazlENUxlZUJaaTEwQVFWekJlQXxBQ3Jtc0tudDJFeXJpRkdGYU83NkcxbW1TdEdnM0hYTmxCLXhEWC1qOHBYVEZGUmVtUmJfb3ZaWTdOWURWTHZjMjMwdnN2SmNGMkxvNlUwdmhfcWRQMnNuZWV5bEZoVlJObzZQN1RLcWNHVFFZOVBjQThvc210SQ&q=http%3A%2F%2Fjonathansoma.com%2Flede%2Fdata-studio%2Fmatplotlib%2Flist-all-fonts-available-in-matplotlib-plus-samples%2F&v=DAQNHzOcO5A

#X and Y labels
plt.xlabel('X Label')
plt.ylabel('Y Label')

#X, Y axis tickmarks (scale of the graph) - make the axis whole/exact numbers and not decimals
plt.xticks([1,2,3])
plt.yticks([0,2,4,6,8])

#placing legends - declared as 'label='2x''
plt.legend()

#save graph (dpi 300 is good when saving so graph has high resolution)
plt.savefig('first graph.png', dpi = 300) #save the image called plot.png

plt.show()
```

### Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]
#plt.bar(labels, values)
bars = plt.bar(labels, values)

#Have patterns in each bar
patterns = ['/', 'o', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
    
#same as above
#bars[0].set_hatch('/')
#bars[1].set_hatch('o')
#bars[2].set_hatch('*')

plt.figure(figsize=(6,4))
plt.show()
```

### Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Line Graph

```python
gas = pd.read_csv('Documents\Python\Plots\Graphs Examples Data\gas_prices.csv')

plt.figure(figsize=(8,5))

plt.title('Gas Prices in USD', fontdict={'fontweight':'bold', 'fontsize':18})

plt.plot(gas.Year, gas.USA, 'b*-', label = 'USA') #column Year is X axis, USA price is Y axis
plt.plot(gas.Year, gas.Canada, 'r.-', label = 'CAN') 
plt.plot(gas['Year'], gas['South Korea'], 'g.-', label= 'SN') #same sintaxis as above
plt.plot(gas['Year'], gas['Australia'], 'y.-', label= 'AU')

plt.xlabel('Year')
plt.ylabel('Price')

plt.xticks(gas['Year'][::3].tolist()+[2011]) #x axis every 3 years and adding the limit until 2011

plt.legend()

plt.show()
```

```python
gas
```

```python
#Plotting all countries
for country in gas:
    if country != 'Year': #selecting all columns that are not Years column
        plt.plot(gas['Year'], gas[country])
```

```python
#Plotting selected countries
countries_to_look_at = ['Australia', 'USA', 'Canada']
for country in gas:
    if country in countries_to_look_at: #selecting countries declared above
        plt.plot(gas['Year'], gas[country])
```

### Load Fifa Data

```python
fifa = pd.read_csv('Documents\Python\Plots\Graphs Examples Data\_fifa_data.csv')

fifa
```

### Histograms

```python
segment = [40,50,60,70,80,90,100]

plt.hist(fifa['Overall'], bins=segment, color='#5aedba') #bins are on x axis, color code can also be looked in google as color picker

plt.xticks(segment) #interval declared the segment list

plt.xlabel('Skill level')
plt.ylabel('Number of Players')
plt.title('Player Skilss in FIFA')

plt.show()
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

etiq = ['Left', 'Right']
ilum = ['#abcdef', '#aabbcc']

plt.pie([left, right], labels = etiq, colors=ilum, autopct='%.2f %%')

plt.title('Foot Preference of FIFA Players')
plt.show()
```
